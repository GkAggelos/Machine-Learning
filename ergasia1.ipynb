2.1 Η παράγωγος της f(x) υπολογίζεται ως εξής :

$$ \frac{\partial f(x)}{\partial x} = \frac{\partial (x^{T}Ax + b^{T}x)}{\partial x} = \frac{\partial x^{T}Ax}{\partial x} + \frac{\partial b^{T}x}{\partial x}$$

Από το matrixcookbook έχω ότι :

$$\frac{dx^{T}Bx}{\partial x} = (B + B^{T})x$$
$$\frac{da^{T}x}{\partial x} = a$$

Άρα έχω :

$$\frac{\partial x^{T}Ax}{\partial x} + \frac{\partial b^{T}x}{\partial x} = (A + A^{T})x + b$$

και αφού ο Α συμμετρικός έχω :

$$(A + A^{T})x + b = 2Ax + b$$

2.2
Έχω f(x)=$||A-XB||^{2}_{F}$

Ξέρουμε ότι η Ευκλείδεια νόρμα είναι κυρτή συνάρτηση άρα το ολικό ελάχιστο θα είναι στο σημείο που μηδενίζεται η παράγωγος της συνάρτησης.

f(x)=$||A-XB||^{2}_{F}=Tr((A-XB)(A-XB)^{T})=Tr((A-XB)(A^{T}-(XB)^{T}))=Tr((A-XB)(A^{T}-B^{T}X^{T}))=Tr((AA^{T})-(XBA^{T})-(AB^{T}X^{T})+(XBB^{T}X^{T}))=
Tr(AA^{T})-Tr(XBA^{T})-Tr(AB^{T}X^{T})+Tr(XBB^{T}X^{T})$

και έχω να λύσω την εξίσωση

$$\frac{\partial f(x)}{\partial x}=0$$

$$\frac{\partial (Tr(AA^{T})-Tr(XBA^{T})-Tr(AB^{T}X^{T})+Tr(XBB^{T}X^{T}))}{\partial x}=0$$

$$\frac{\partial Tr(AA^{T})}{\partial x}-\frac{\partial Tr(XBA^{T})}{\partial x}-\frac{\partial Tr(AB^{T}X^{T})}{\partial x}+\frac{\partial Tr(XBB^{T}X^{T})}{\partial x}=0$$

Από το matrixcookbook έχω ότι :

$$\frac{\partial Tr(AA^{T})}{\partial x}=0$$
$$\frac{\partial Tr(XBA^{T})}{\partial x}=(BA^{T})^{T}$$
$$\frac{\partial Tr(AB^{T}X^{T})}{\partial x}=AB^{T}$$
$$\frac{\partial Tr(XBB^{T}X^{T})}{\partial x}=X(BB^{T})^{T}+XBB^{T}$$

Άρα έχω

$$0-(BA^{T})^{T}-AB^{T}+X(BB^{T})^{T}+XBB^{T}=0$$

$$-AB^{T}-AB^{T}+XBB^{T}+XBB^{T}=0$$

$$-2AB^{T}+2XBB^{T}=0$$

$$2XBB^{T}=2AB^{T}$$

$$XBB^{T}=AB^{T}$$

$$X=AB^{T}(BB^{T})^{-1}$$

```python
import numpy as np
import matplotlib.pyplot as plt

def power(a):
  return (a*a)

def df1X1(x):
  return (2*(x-2))

def df1X2(x):
  return (2*(x-3))

def f1(x1,x2):
  return (power(x1 - 2) + power(x2 - 3))


def gradient_descent_f1(curX1,curX2,maxIters,rate,p) :
  ValueOfF1 =[]
  Iterations1 = []
  plt.figure()
  precision = 0.001

  for aIter in range(maxIters):
    prevX1=curX1
    prevX2=curX2
    curX1=curX1-rate*df1X1(curX1)
    curX2=curX2-rate*df1X2(curX2)
    diff1=abs(curX1-prevX1)
    diff2=abs(curX2-prevX2)
    if diff1<=precision and diff2<=precision:
      print("The local minimum occurs at", (curX1,curX2))
      break
    ValueOfF1.append(f1(curX1,curX2))
    Iterations1.append(aIter)
    if(p):
      print("Iteration",aIter,"\nf1(",curX1,",",curX2,") =",f1(curX1,curX2)) #Print iterations

  plt.scatter(Iterations1,ValueOfF1)
  plt.xlabel("Iterations")
  plt.ylabel("F1")
  plt.show()



def df2X1(x1,x2):
  return (40*((x1 + 3) - power(x2 - 3)))

def df2X2(x1,x2):
  return ((-2 * (1 - (x2 - 3))) - (80*((x1 + 3) - power(x2 - 3)) * (x2 - 3)))

def f2(x1,x2):
  return (power(1 - (x2 - 3)) + 20*power((x1 + 3) - power(x2 - 3)))

#rate = 0.005
#maxIters = 1000
def gradient_descent_f2(curX1,curX2,maxIters,rate,p) :
  ValueOfF2 =[]
  Iterations2 = []
  plt.figure()
  precision = 0.001

  for aIter in range(maxIters):
    prevX1 = curX1
    prevX2 = curX2
    curX1=curX1-rate*df2X1(curX1,curX2)
    curX2=curX2-rate*df2X2(curX1,curX2)
    diff1=abs(curX1-prevX1)
    diff2=abs(curX2-prevX2)
    if diff1<=precision and diff2<=precision:
      print("The local minimum occurs at", (curX1,curX2))
      break
    ValueOfF2.append(f2(curX1,curX2))
    Iterations2.append(aIter)
    if(p):
      print("Iteration",aIter,"\nf2(",curX1,",",curX2,") =",f2(curX1,curX2)) #Print iterations

  plt.scatter(Iterations2,ValueOfF2)
  plt.xlabel("Iterations")
  plt.ylabel("F2")
  plt.show()
```

##Παρτηρήσεις για την f1

Στην f1 για maxIters=10 και rate=0.5 χρειάζεται μόνο μία επανάληψη για να συγκλίνει ο αλγόριθμος στο τοπικό ελάχιστο. Για maxIters=100 και rate=0.1 χρειάζεται 29 επαναλήψεις. Για maxIters=100 και rate=0.05 χρειάζεται 55 επαναλήψεις. Για maxIters=1000 και rate=0.01 χρειάζεται πάνω από 100 επαναλήψεις(203). Έτσι διαπιστώνουμε ότι για μικρές τιμές του rate αργεί περισσότερο να βρεί την λύση σε σχέση με μεγάλες τιμές, ενώ επίσης πέφτει και η ακρίβεια καθώς για rate=0.5 έχω x1=2, x2=3, ενώ για rate=0.05 εχω x1=1.9945221451000934, x2=2.9917832176501395 και για rate=0.01 εχω x1=1.967554876102205, x2=2.951332314153307. Για μεγαλύτερες αρχικές τιμές βλέπω ότι με rate=0.5 θέλει πάλι μία επανάληψη, ενώ στα πιο μικρά rate θέλει παραπάνω.Τέλος για τιμές πάνω του 0.5 αργεί να συγκλίνει όσο προσεγγίζουμε το 1 και όταν το ξεπεράσουμε δεν συγκλινει ποτέ.

```python
gradient_descent_f1(0.0,0.0,10,0.5,1)
gradient_descent_f1(0.0,0.0,100,0.1,1)
gradient_descent_f1(0.0,0.0,100,0.05,0)
gradient_descent_f1(0.0,0.0,1000,0.01,0)
gradient_descent_f1(10.0,20.0,10,0.5,0)
gradient_descent_f1(10.0,20.0,100,0.1,0)
gradient_descent_f1(10.0,20.0,100,0.05,0)
gradient_descent_f1(10.0,20.0,1000,0.01,0)
gradient_descent_f1(0.0,0.0,1000,0.8,0)
gradient_descent_f1(0.0,0.0,1000,1,0)
```
